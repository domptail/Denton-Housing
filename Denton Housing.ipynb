## 0. Import the data and visualize it

```python
# Look at the raw data
import pandas as pd
df = pd.read_csv('denton_housing.csv')
df
```

```python
# Delete the unnamed columns for more lisibility
del df['Unnamed: 3']
del df['Unnamed: 4']
del df['Unnamed: 5']
del df['Unnamed: 6']
df
```

## 1. Calculate the percentage increase or decrease in 'total housing units' from year to year

```python
# Create a new dataframe with only the data on 'Total Housing Units'
df_total = df[(df['title_field']) == 'Total Housing Units']

# Reorder the data from 2008 to 2014
df_total = df_total.sort_values(by = 'year', ascending = 1)

df_total
```

```python
# METHOD 1: Calculate evolution from year to year using the pct_change() method
evolution1 = pd.DataFrame()
evolution1['Years'] = df_total['year']
evolution1['Increase or decrease (%)'] = round((df_total['value'].pct_change())*100, 1)
print(evolution1)
```

```python
# METHOD 2: Calculate evolution from year to year 'manually'
change = []

# Cannot calculate evolution for the first year (2008)
change.append('NaN')

# Calculate evolution for each subsequent year with a 'for' loop
for i in range(1,7):
    current_value = df_total.iloc[i]['value']
    previous_value = df_total.iloc[i-1]['value']
    change_val = (current_value - previous_value) / previous_value
    change_pct = round(change_val*100, 1)
    change.append(change_pct)

evolution2 = pd.DataFrame()
evolution2['Years'] = df_total['year']   
evolution2['Increase or decrease (%)'] = change
    
print(evolution2)    
```

## 2. Calculate descriptive statistics on 'vacant housing units'

```python
# Create a new dataframe with only the data on 'Vacant Housing Units'
df_vacant = df[(df['title_field']) == 'Vacant Housing Units']
df_vacant
```

```python
# A) Average number of vacant housing units
vacant_avg = df_vacant['value'].mean()
print(vacant_avg)
```

```python
# B) Median of vacant housing units
vacant_median = df_vacant['value'].median()
print(vacant_median)
```

```python
# C) Range of vacant housing units
vacant_range = df_vacant['value'].max() - df_vacant['value'].min()
print(vacant_range)
```

```python
# D) Standard deviation of vacant housing units
vacant_std = df_vacant['value'].std()
print(vacant_std)
```

```python
# Alternatively can use the describe method
df_vacant['value'].describe()
```

## 3. Visualize the data

```python
# Create a new dataframe with only the data on 'Occupied Housing Units'
df_occupied = df[(df['title_field']) == 'Occupied Housing Units']

# Import library
import matplotlib.pyplot as plt

# Plot Vacant, Occupied and Total Housing Units
plt.plot(df_total['year'], df_total['value'], label = 'Total Housing Units')
plt.legend(loc='right')
plt.xlabel('Year')
plt.ylabel('Number of Units')

plt.plot(df_occupied['year'], df_occupied['value'], label = 'Occupied Housing Units')
plt.legend(loc='right')

plt.plot(df_vacant['year'], df_vacant['value'], label = 'Vacant Housing Units')
plt.legend(loc='right')

plt.show()
```

```python
# Plot Owner and Renter Occupied Units
df_renter = df[(df['title_field']) == 'Renter Occupied']

df_owner = df[(df['title_field']) == 'Owner Occupied']

plt.plot(df_renter['year'], df_renter['value'], label = 'Renter Occupied')
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('Number of Units')

plt.plot(df_owner['year'], df_owner['value'], label = 'Owner Occupied')
plt.legend(loc='upper left')

plt.show()
```
